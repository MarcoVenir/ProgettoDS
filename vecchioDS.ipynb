# Generazione delle build casuali 

Funziona utilizzata... campi e parametri...
Slides: Numpy2 

```python
import numpy as np 
import pandas as pd

dati = []
n_righe = 5000
colonne = ['VITALITÀ', 'ARMONIZZAZIONE', 'TEMPRA', 'VIGORE', 'FORZA', 'DESTREZZA', 'INTELLIGENZA', 'FEDE', 'FORTUNA']

def genera_riga():
    while True:
        valori = np.random.randint(1, 50, size=len(colonne))
        if valori.sum() <= 150:
            return list(valori)

for _ in range(n_righe):
    dati.append(genera_riga())
    

build = pd.DataFrame(dati, columns=colonne) 

display(build) 
```
